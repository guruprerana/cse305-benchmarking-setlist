```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_coarse = pd.read_csv("benchmarking_coarse.csv", index_col=0)
data_fine = pd.read_csv("benchmarking_fine.csv", index_col=0)
```

```python
data_coarse
```

```python
data_fine
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 10))

for i in range(1, 9):
    ax.plot(data_coarse.columns, data_coarse.loc[i], label=f"{i} threads")

ax.legend()
ax.set_xlabel("Number of insertions")
ax.set_ylabel("Time (ms)")
fig.suptitle("Benchmarking coarse-grained insertion", fontsize=16)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 10))

for i in range(1, 9):
    ax.plot(data_fine.columns, data_fine.loc[i], label=f"{i} threads")

ax.legend()
ax.set_xlabel("Number of insertions")
ax.set_ylabel("Time (ms)")
fig.suptitle("Benchmarking fine-grained insertion", fontsize=16)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 10))

ax1.plot(data_coarse.columns, data_coarse.loc[1], label="coarse-grained")
ax1.plot(data_fine.columns, data_fine.loc[1], label="fine-grained")
ax1.legend()
ax1.title.set_text("1 thread")

ax2.plot(data_coarse.columns, data_coarse.loc[3], label="coarse-grained")
ax2.plot(data_fine.columns, data_fine.loc[3], label="fine-grained")
ax2.legend()
ax2.title.set_text("3 thread")

ax3.plot(data_coarse.columns, data_coarse.loc[6], label="coarse-grained")
ax3.plot(data_fine.columns, data_fine.loc[6], label="fine-grained")
ax3.legend()
ax3.title.set_text("6 thread")

ax4.plot(data_coarse.columns, data_coarse.loc[8], label="coarse-grained")
ax4.plot(data_fine.columns, data_fine.loc[8], label="fine-grained")
ax4.legend()
ax4.title.set_text("8 threads")

fig.suptitle("Coarse-grained vs. Fine-grained", fontsize=16)
```
